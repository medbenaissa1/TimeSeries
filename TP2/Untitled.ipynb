1-	Une série prédictible est une série temporelle dont les valeurs futures peuvent être estimées à partir de ses valeurs passées. Une série non prédictible est une série qui ne présente pas difficile à faire des prévisions précises


2-	Une série temporelle est composée de : saisonalité,tendance,cycle et résidual.


3-	Vérifier la stationnarité pour effectuer des prévisions précises qui seront manipulés ensuite dans ARIMA


4-	Il faut que la moyenne et la variance ne varie pas et pas d’auto-corrélation


5-	Les graphiques : ADF, ACF


6-	Permet de confirmer la stationnarité en effectuant des tests d’hypothèses


```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000  

bruit_blanc = np.random.normal(loc=0.0, scale=1.0, size=n)


plt.figure(figsize=(10, 5))
plt.plot(bruit_blanc, label='Bruit blanc')
plt.axhline(y=np.mean(bruit_blanc) , color = "yellow"  )
plt.title('Série de Bruit Blanc')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
# statistical test
result = adfuller(bruit_blanc)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

Puisque P-value est nulle , il est confirmée que white noise est stationnaire


Random Walk


```python
import random
import matplotlib.pyplot as plt

random.seed(1)

random_walk = []

# Start the random walk at either -1 or 
random_walk.append(-1 if random.random() < 0.5 else 1)

# Generate the random walk series
for i in range(1, 1000):
    movement = -1 if random.random() < 0.5 else 1
    value = random_walk[i-1] + movement
    random_walk.append(value)

# Plot the random walk
plt.figure(figsize=(10, 5))
plt.plot(random_walk, label="Random Walk")
plt.title("Random Walk Series")
plt.xlabel("Steps")
plt.ylabel("Value")
plt.legend()
plt.show()
```
