*Bruit Blanc *

---



```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000

bruit_blanc = np.random.normal(loc=0.0, scale=1.0, size=n)


plt.figure(figsize=(10, 5))
plt.plot(bruit_blanc, label='Bruit blanc')
plt.axhline(np.mean(bruit_blanc) , color = "yellow"  )
plt.title('Série de Bruit Blanc')
plt.show()
```

Vérifier la stationnarité de la série générée en utilisant ADF (On peut constater que la moyenne et la variance sont stable)

```python
from statsmodels.tsa.stattools import adfuller
# statistical test
result = adfuller(bruit_blanc)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

Puisque P-value est nulle , il est confirmée que white noise est stationnaire.


Une série de bruit blanc est non-prédictible, car il n’y a aucune structure ou tendance dans les données qui pourrait aider à prévoir les valeurs futures.

*Random walk*

---



```python
import random
import matplotlib.pyplot as plt


random.seed(1)

random_walk = []

random_walk.append(-1 if random.random() < 0.5 else 1)

for i in range(1, 1000):
    movement = -1 if random.random() < 0.5 else 1
    value = random_walk[i-1] + movement
    random_walk.append(value)

plt.figure(figsize=(10, 5))
plt.plot(random_walk, label="Random Walk")
plt.title("Random Walk Series")
plt.xlabel("Steps")
plt.ylabel("Value")
plt.legend()
plt.show()
```

La moyenne et la variance ne sont pas stable donc random walk n'est pas stationnaire . On confirme avec ADF :

```python
from statsmodels.tsa.stattools import adfuller
# statistical test
result = adfuller(random_walk)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

p-value > 0.05  . Donc random_walk effectivement n'est pas stationnaire.

La valeur prochaine dépend de la valeur courante, et il y'a une tendance croissante. Donc Random Walk est prédictible.




**Exercice 3: Trend et saisonalité**

---



Un modèle additif s'écrit sous la forme:

*   y(t) = Trend + Saisonalité + Bruit
*   Il est linéaire : Même qte de changement en fct du temps (Trend et Saisonalité Sont linéaires).



Un modèle multiplicatif s'écrit sous la forme:


*   y(t) = Trend * Saisonalité * Bruit
*   Il n'est pas linéaire: Qte de changement différents en fct du temps(Trend et saisonalité non linéaires).



**Série avec Trend (modèle aditif)**

---



```python
np.random.seed(0)

n_points = 100
pente_trend = 0.5


time = np.arange(n_points)
trend = time * pente_trend
bruit = np.random.normal(0,1,n_points)

series= trend + bruit
plt.figure(figsize=(10, 5))
plt.plot(time,series, label='Série avec Trend')
plt.plot(time,trend, color="orange", label="Trend")
plt.legend()
plt.title('Série avec Trend(modèle additif)')
plt.show()

```

La série n'est pas stationnaire : il y a la composante trend et la moyenne et la variance changent en fct du temps.On vérifie avec ADF:

```python
from statsmodels.tsa.stattools import adfuller
# statistical test
result = adfuller(series)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

p-value>0.05. donc effectivement la série n'est pas stationnaire.

On supprime la tendance pour avoir la stationnarité, Pour améliorer notre modélisation et rendre les prévisions plus correcte. On utilise:

*   La différenciation
*   Régression ou Auto-regression
*   Décomposition







```python
##Differenciation

diff = []

for i in range(1,len(series)):
  value = series[i]-series[i-1]
  diff.append(value)

plt.figure(figsize=(10, 5))
# Original series
plt.subplot(1, 2, 1)
plt.plot(time, series, label='Série Temporelle', color='blue')
plt.title('Série Temporelle avec Tendance')
plt.xlabel('Temps')
plt.ylabel('Valeur')
plt.legend()

# Differentiated series
plt.subplot(1, 2, 2)
plt.plot(time[1:], diff, label='Série Différenciée', color='green')
plt.title('Série Différenciée')
plt.xlabel('Temps')
plt.ylabel('Différence')
plt.legend()


plt.show()
```

```python
#fit linear model

X = [i for i in range(0, len(series))]
X = np.reshape(X, (len(X), 1))
y = series
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
trend_fit = model.predict(X)

plt.figure(figsize=(12, 5))
plt.plot(time, y, label='Série Temporelle', color='blue')
plt.plot(time, trend_fit, label='Tendance Ajustée', color='orange', linestyle='--')
plt.title('Série Temporelle avec Tendance Ajustée')
plt.xlabel('Temps')
plt.ylabel('Valeur')
plt.legend()
plt.show()


```

```python
#detrend

detrended = [y[i]-trend_fit[i] for i in range(0, len(series))]

plt.figure(figsize=(10, 5))
plt.plot(time, detrended, label='Série Détrendée', color='red')
plt.legend()
plt.show()
```

**Série avec saisonnalité (modèle aditif)**

---



```python

np.random.seed(42)

n_periods = 120
seasonal_period = 12
frequency = np.arange(n_periods)

seasonal_component = 10 * np.cos(2 * np.pi * frequency / seasonal_period)

noise = 2*np.random.normal(0, 2, n_periods)

time_series = seasonal_component + noise

# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(time_series, label='Time Series with Seasonality and Noise')
plt.axhline(y=np.mean(time_series), color='r', linestyle='--', label='Mean')
plt.axhline(y=np.mean(time_series) + np.std(time_series), color='g', linestyle='--', label='Mean + 1 Std Dev')
plt.axhline(y=np.mean(time_series) - np.std(time_series), color='g', linestyle='--', label='Mean - 1 Std Dev')
plt.title('Generated Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

La série n'est pas stationnaire.


*    La ligne moyenne  est approximativement centrée autour de zéro.
*   Une certaine stabilité de la variance
*   Effectuant ADF:










```python
from statsmodels.tsa.stattools import adfuller
# statistical test
result = adfuller(time_series)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))


```

0.7% indique une stationnarité du modèle.

```python
#differenciating

diff = time_series[12:] - time_series[:-12]
plt.figure(figsize=(12, 6))
plt.plot(diff, label="Série après differencing saisonnier")
plt.xlabel("Temps")
plt.ylabel("Valeur différenciée")
plt.title("Série après differencing saisonnier")
plt.legend()
plt.show()
```

```python
#Décomposition

from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(time_series, model='additive', period=12)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(12, 10))
plt.subplot(4, 1, 1)
plt.plot(time_series, label="Série originale")
plt.legend(loc="upper left")
plt.subplot(4, 1, 2)
plt.plot(trend, label="Composante de Tendance", color='orange')
plt.legend(loc="upper left")
plt.subplot(4, 1, 3)
plt.plot(seasonal, label="Composante Saisonnalité", color='green')
plt.legend(loc="upper left")
plt.subplot(4, 1, 4)
plt.plot(residual, label="Résidu (Bruit)", color='red')
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()


```

**Série avec Trend et saisonalité(modèle additif)**

---



```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(0)
n = 200

trend = 0.5 * np.arange(n)

period = 50
amplitude = 10
seasonality = amplitude * np.sin(2 * np.pi * np.arange(n) / period)

noise = np.random.normal(0, 2, n)  # Bruit avec moyenne 0 et écart-type 2

time_series = trend + seasonality + noise

# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(time_series, label="Série Temporelle")
plt.plot(trend, label="Tendance (Trend)", linestyle="--",color="red")
plt.plot(seasonality, label="Saisonnalité (Seasonality)", linestyle="--",color="green")
plt.plot(noise, label="Bruit Blanc", linestyle="--",color="yellow")
plt.title("Série Temporelle avec Tendance et Saisonnalité(Modèle additif)")
plt.xlabel("Temps")
plt.ylabel("Valeur")
plt.legend()
plt.show()

```

Il ya la composante trend et saisonalité. La série n'est pas stationnaire.
La moyenne varie ! On vérifie avec ADF:

```python
from statsmodels.tsa.stattools import adfuller
# statistical test
result = adfuller(time_series)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

Effectivement p-value est 95% > 50% . Donc la série n'est pas stationnaire.

```python
#Décomposition

from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(time_series, model='additive', period=period)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(12, 10))
plt.subplot(4, 1, 1)
plt.plot(time_series, label="Série originale")
plt.legend(loc="upper left")
plt.subplot(4, 1, 2)
plt.plot(trend, label="Composante de Tendance", color='orange')
plt.legend(loc="upper left")
plt.subplot(4, 1, 3)
plt.plot(seasonal, label="Composante Saisonnalité", color='green')
plt.legend(loc="upper left")
plt.subplot(4, 1, 4)
plt.plot(residual, label="Résidu (Bruit)", color='red')
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
#Detrended

detrended_series = time_series - trend
without_trend_and_seasonality = detrended_series - seasonal

plt.figure(figsize=(12, 10))

plt.subplot(3, 1, 1)
plt.plot(time_series, label="Série originale")
plt.legend(loc="upper left")

plt.subplot(3, 1, 2)
plt.plot(detrended_series, label="Sans tendance", color='orange')
plt.legend(loc="upper left")

plt.subplot(3, 1, 3)
plt.plot(without_trend_and_seasonality, label="Sans tendance et saisonnalité", color='green')
plt.legend(loc="upper left")

plt.tight_layout()
plt.show()
```

**Série avec Trend & saisonnalité (modèle multiplicatif)**

```python


np.random.seed(0)
n = 200

slope = 0.01
trend = 1 + slope * np.arange(n)

period = 50
amplitude = 0.5
seasonality = 1 + amplitude * np.sin(2 * np.pi * np.arange(n) / period)


noise = np.random.normal(1, 0.5, n)


time_series = trend * seasonality * noise


plt.figure(figsize=(12, 8))

plt.plot(trend, label="Tendance (Trend)", color="blue")
plt.plot(seasonality, label="Saisonnalité (Seasonality)", color="orange")
plt.plot(noise, label="Bruit Blanc (White Noise)", color="green")
plt.plot(time_series, label="Série Temporelle (Multiplicative)", color="purple")
plt.legend()
plt.title("Série Temporelle avec Tendance et Saisonnalité (Modèle multiplicatif)")
plt.xlabel("Temps")
plt.ylabel("Valeur")
plt.show()

```

On vérifie avec ADF la stationnarité:

```python
from statsmodels.tsa.stattools import adfuller
# statistical test
result = adfuller(time_series)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

P-value = 1.9% ! Donc la série est stationnaire
