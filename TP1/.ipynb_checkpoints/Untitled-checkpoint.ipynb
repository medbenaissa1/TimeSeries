Basic Feature Engineering

```python

temps = pd.DataFrame(series)
Sales_lag1 = temps.shift(1)
Sales_rolling_7 = Sales_lag1.rolling(window = 7)
means = Sales_rolling_7.mean()
print(means.head(10))
```

```python

temps.index
```

```python
def extract(x):
    year,month,day = x.split("-")
    return year,month,day
    
    
```

```python
years =[]
months = []
days = []
for i in range(len(temps.index)):
    year,month,day = extract(str(temps.index[i]))
    years.append(year)
    months.append(month)
    days.append(day)
```

```python
months
```

```python
temps["Mois"] = months
temps["Days"] = days
temps["Days"] = temps["Days"].astype(str).str.split(" ").str[0]
```

```python
temps
```

```python
#Resampling
monthly_sales = series['Sales'].resample('M').sum()
print(monthly_sales.head())
```

```python
#interpolation
series['Sales'].replace(0, pd.NA, inplace=True)
series['Sales'] = series['Sales'].interpolate(method='linear')
print(series.head())
```
