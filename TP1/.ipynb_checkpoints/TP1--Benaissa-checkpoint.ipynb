```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Basic Feature Engineering

```python

temps = pd.DataFrame(series)
Sales_lag1 = temps.shift(1)
Sales_rolling_7 = Sales_lag1.rolling(window = 7)
means = Sales_rolling_7.mean()
print(means.head(10))
```

```python

temps.index
```

```python
def extract(x):
    year,month,day = x.split("-")
    return year,month,day
    
    
```

```python
years =[]
months = []
days = []
for i in range(len(temps.index)):
    year,month,day = extract(str(temps.index[i]))
    years.append(year)
    months.append(month)
    days.append(day)
```

```python
months
```

```python
temps["Mois"] = months
temps["Days"] = days
temps["Days"] = temps["Days"].astype(str).str.split(" ").str[0]
```

```python
temps
```

```python
#Resampling et Interpolation
```

```python
series
```

```python
#Resampling
monthly_sales = series['Sales'].resample('M').sum()
print(monthly_sales.head())
```

```python
#interpolation
series['Sales'].replace(0, pd.NA, inplace=True)
series['Sales'] = series['Sales'].interpolate(method='linear')
print(series.head())
```

Transformers


```python
#Log transformer


log_sales = np.log(series)
#avant
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(series, label='Sales (Original)')
plt.title('Original Sales Data')
plt.legend()

#aprÃ¨s
plt.subplot(2, 1, 2)
plt.plot(log_sales, label='Log(Sales)', color='green')
plt.title('Sales Data after Log Transformation')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Cox Box Transformer
from scipy import stats

boxcox_sales,lambda_value = stats.boxcox(series["Sales"])

plt.figure(figsize=(10, 5))

# SÃ©rie originale
plt.subplot(2, 1, 1)
plt.plot(series, label='Sales (Original)')
plt.title('Original Sales Data')
plt.legend()

# SÃ©rie aprÃ¨s transformation Box-Cox
plt.subplot(2, 1, 2)
plt.plot(boxcox_sales, label=f'Box-Cox Sales (lambda={lambda_value:.2f})', color='Red')
plt.title('Sales Data after Box-Cox Transformation')
plt.legend()

plt.tight_layout()
plt.show()

# Afficher la valeur du lambda choisi pour la transformation Box-Cox
print(f"Box-Cox Lambda Value: {lambda_value}")
```
