*INTERPOLATION VS EXTRAPOLATION*






> ***INTERPOLATION***


*  **Définition** : L'interpolation consiste à estimer des valeurs à l'intérieur de l'intervalle de données existant. Elle utilise les points connus pour prédire des valeurs situées entre ces points.
*  **Utilisation** : Elle est souvent utilisée quand on a des valeurs de données manquantes ou qu'on souhaite lisser les données.
*  **Exemple** : Si on a des données de températures relevées toutes les heures de 10 h à 18 h, et qu’on souhaite estimer la température à 10 h 30, on utilisera l’interpolation.




> ***EXTRAPOLATION***


*   **Définition** : L'extrapolation consiste à estimer des valeurs situées en dehors de l'intervalle de données existant. Elle utilise une tendance observée dans les points existants pour prédire des valeurs au-delà de la plage connue.
*   **Utilisation** : Elle est souvent utilisée pour faire des prévisions ou anticiper des valeurs futures.
*   **Exemple** : Avec les mêmes relevés de températures de 10 h à 18 h, si on souhaite estimer la température à 19 h ou 9 h, on utilisera l’extrapolation.

Les constituants d'une série temporelle:

1.   Trend (optionnel): Croissance ou décroissance continue de la série temporelle par rapport au temps.
2.   Seasonality (optionnel): un pattern qui se répète dans le temps.
3. Cyclic: fait référence aux fluctuations (régulières ou périodiques) autour de la tendance.
4. Noise : variations dans les observation que le modèle ne peut pas expliquer.

*   Une taille importante nous permet de séparer le dataset en dataset d’apprentissage, de test et de validation.







Quelle est la qualité du dataset?

Pour quel horizon les predictions sont juducieuses?


*   Court, moyen ou à long terme? Les prévisions à court terme sont plus juducieuses.
